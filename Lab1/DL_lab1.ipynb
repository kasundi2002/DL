**Histogram**
Purpose: **Shows how data is distributed across different value ranges** (bins).

How it works:

It **splits your data into intervals** (called bins).

Counts **how many values fall into each bin**.

Displays this as bars — taller bars mean more values in that range.

**Heatmap**
Purpose: **Displays values in a matrix using colors to represent magnitude.**

How it works:

You have a **2D table of numbers** (like a correlation matrix).

**Each cell’s color intensity corresponds to the value.**

Often **includes annotations** (numbers written inside the cells).

2. Create a random 100000x1 array from the same exponential distribution and visualize the
distribution histogram along with uniform and normal distributions.

```python
import matplotlib.pyplot as plt
```

```python
#Generates 100,000 values from an exponential distribution.
exp_data = np.random.exponential(scale=1.0, size=100000)
```

```python
#Generates 100,000 values from a uniform distribution (between 0 and 1).
uniform_data = np.random.rand(100000)
```

```python
#Generates 100,000 values from a uniform distribution (between 0 and 1).
normal_data = np.random.randn(100000)
```

```python
#Plots the exponential distribution as a histogram.
#density=True → normalizes the histogram to a probability density.
#bins=100 → number of intervals.
#histtype="step" → draws just the outline of the bars.
#label="Exponential" → legend label.

plt.hist(exp_data,density=True,bins=100,histtype="step",color="red",label="exponential")
plt.hist(uniform_data,density=True,bins=100,histtype="step",color="blue",label="Uniform")

#Sets axis limits: x from -2.5 to 7.5, y from 0 to 1.1.
plt.axis([-2.5,7.5,0,1.1])
plt.hist(normal_data,density=True,bins=100,histtype="step",color="green",label="Normal")
#Shows the legend in the top-right corner.
plt.legend(loc="upper right")
plt.title("Rendom Distributions")
plt.xlabel("Value")
plt.ylabel("Density")
plt.show()
```

 1.3 3D plot of Z = X² + Y²

```python
#Imports 3D plotting tools from Matplotlib.
from mpl_toolkits.mplot3d import Axes3D
#Creates 100 evenly spaced values from -5 to 5 for x and y
x = np.linspace(-5,5,100)
y = np.linspace(-5,5,100)
X,Y = np.meshgrid(x,y)
#Creates a grid of x and y values.
Z = X**2 + Y**2
#Calculates z for each (x, y) point using the formula x² + y².
fig = plt.figure()
ax = fig.add_subplot(111,projection='3d')
ax.plot_surface(X,Y,Z, cmap='viridis')
#Creates a 3D surface plot with a color map.
ax.set_title("Z = X^2 + Y^2")
plt.show()
```

1.4 Pearson & Spearman correlation of Pokémon features

```python
# Import libraries
#Imports Pandas for data handling and Seaborn for visualization.
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt

# Load the dataset
!wget -q https://elitedatascience.com/wp-content/uploads/2022/07/Pokemon.csv
df = pd.read_csv('Pokemon.csv', index_col=0, encoding='latin')

# Select relevant numerical features
stats_df = df[['HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed']]

#Calculates correlation matrices for Pearson (linear) and Spearman (rank).
# Pearson Correlation
pearson_corr = stats_df.corr(method='pearson')
print("Pearson Correlation Matrix:")
print(pearson_corr)


# Spearman Correlation
spearman_corr = stats_df.corr(method='spearman')
print("\nSpearman Correlation Matrix:")
print(spearman_corr)

# Plot Pearson heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(pearson_corr)
plt.title('Pearson Correlation Heatmap')
plt.show()

# Plot Spearman heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(spearman_corr)
plt.title('Spearman Correlation Heatmap')
plt.show()
```

```python
df = pd.read_csv("Au_nanoParticle_dataset.csv")
```

Loads CSV data into a Pandas DataFrame.

```python
new_df = df[['N_total', 'N_bulk', 'N_surface', 'R_avg']]
```

Creates a new DataFrame with only the required columns.

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(16, 4))
for i, column in enumerate(new_df.columns):
    plt.subplot(1, 4, i + 1)
    plt.hist(new_df[column], bins=20, color='skyblue')
    plt.title(column)
    plt.xlabel("Value")
    plt.ylabel("Frequency")
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns

sns.pairplot(new_df, diag_kind='hist')
```

```python
import seaborn as sns

g = sns.PairGrid(new_df)
g.map_upper(sns.histplot)
g.map_diag(sns.histplot, kde=True)
g.map_lower(sns.kdeplot, fill=True)
```
